```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('happiness_2017.csv')

# Data Cleaning for Plotting
df_clean = df.dropna(subset=['Region', 'HappinessScore'])

# 1. Histogram: Distribution of Happiness Score
plt.figure(figsize=(10, 6))
sns.histplot(df_clean['HappinessScore'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Happiness Score (2017)')
plt.savefig('happiness_distribution.png')
```

```python
# 2. Box Plot: Happiness Score by Region
plt.figure(figsize=(12, 8))
sns.boxplot(x='HappinessScore', y='Region', data=df_clean, palette='Set3')
plt.title('Happiness Score Distribution by Region')
plt.savefig('happiness_by_region_box.png')
```

```python
# 3. Scatter Plot: Happiness Score vs Log GDP per capita
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='HappinessScore', hue='Region', data=df_clean)
plt.title('Happiness Score vs Log GDP per capita')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('happiness_vs_gdp_scatter.png')
```

```python
# 4. Correlation Heatmap
plt.figure(figsize=(12, 10))
corr = df_clean.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')
```

```python
# 5. Bar Chart: Average Happiness Score by Region
avg_happiness = df_clean.groupby('Region')['HappinessScore'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='HappinessScore', y='Region', data=avg_happiness, palette='viridis')
plt.title('Average Happiness Score by Region')
plt.savefig('avg_happiness_by_region_bar.png')
```

```python
# 6. Pie Chart: Percentage of Countries by Region
region_counts = df_clean['Region'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Countries by Region')
plt.savefig('region_distribution_pie.png')
```
